## Regressão Linear

Uma regressão linear simples consiste em basicamente ajustar uma reta linear que melhor se adequa aos dados. Existem diversos metodos para se criar uma regressão linear, porém todos tem o mesmo objetivo, que é gerar uma função linear do tipo: 


<center>$y = f(X, β)$ </center>

sendo X o conjunto de variáveis independentes e Y a variavel dependente. Os parâmetros desconhecidos e que serão encontrados são β, ou coeficientes da regressão.
</br>

Como mencionado, existem diversos métodos para encontrar uma melhor reta, mas todos seguem o mesmo principio, que é minimizar o erro quadrático atráves dos parâmetros gerados. Alguns famosos são o gradiente descent e o OLS. O método utilizado no trabalho foi o OLS. 

## Random Forest

O segundo método utilizado foi o random forest. Essa técnica é amplamente utilizada dada seu desempenho satisfatório em diversos cenários. O random forest é um "aprimoramento" das árvores de decisão e consiste em basicamente criar, atráves de estratégias coerentes, diversas árvores de decisão para conseguir resultados ainda melhores. A previsão para os dados pode ser a méda encontradas por todas as diferentes árvores criadas. 

A ideia básica do método consiste na realização de um <i>bootstraping</i> nos dados, onde diversas amostras de dados são criadas e para cada uma dessas, uma árvore de decisão é criada usando conjunto de variáveis aleatórios a cada criação de um novo nó da árvore. Isso faz com que árvores diversificadas sejam geradas. Cabe salientar que os nós das árvores são criados de maneira que melhor dividam os dados nas regras no formato se, então. Para a regressão, uma métrica de qualidade pode ser o MSE.  

## LASSO

O último algoritmo utilizado foi o Lasso. A abordagem não se difere muito da regressão linear, onde o objetivo ainda continua encontrar os parâmetros β, chamados coeficientes da regressão. A novidade do modelo fica por conta de uma penalidade adicionada no momento da minimização da erro quadrático para encontrar os coeficientes. 

No Lasso a penalidade adicionada é uma norma L1. Ela é basicamente uma soma das magnitudes dos coeficientes. Esse fator adicionada a minimização faz com que coeficientes menores sejam gerados, reduzindo também a multicolinearidade entre eles.

## Introdução

Com os dados dispostos, a próxima etapa no processo de experimentação a ser realizado consiste na aplicação de um teste para verificar se existe alguma difereça estatisticamente signitiva entre os diferentes modelos aplicados em cada conjunto de dados.

Para a aplicação do teste fez-se necessário a pesquisa quanto a uma alternativa a ser utilizada, visto que a ANOVA aprendida na disciplina não poderia ser utilizada. O motivo para isso baseia-se na violação do pressuposto da independência das unidades experimentais em cada tratamento. No caso do experimento proposto, os conjuntos de dados seriam aplicados em cada um dos modelos.

Considerando o descrito, fez-se uso de uma variação do teste da ANOVA, a ANOVA com medidas repetidas, para verificar a hipótese estabelecida.

A ANOVA com medidas repetidas é um teste voltado para cenários quando mesmas unidades experimentais são expostas a diferentes tratamentos. Devido a isso, há um novo fator na soma de quadrados que precisa ser levado em consideração para que a estatística obtida pelo teste tenha alguma credibilidade.

De modo semelhante a ANOVA independente, há alguns pressupostos que necessitam ser atendidos para que a aplicação da ANOVA com medidas repetidas seja confiável.
Os mais importantes são:
- Normalidade;
- Sphericity.

O primeira diz respeito a necessidade de que os dados oriundos de cada grupo experimental atestem para normalidade. Já quando ao segundo pressuposto, trata-se de algo análogo a homogeneidade das variâncias para a ANOVA independente. Neste caso, trata-se da condição na qual as variâncias da diferenças entre todas as combinações dos grupos relacionados são iguais.
A violação deste último pressusposto, o qual a ANOVA com medidas repetidas é suscetível, induz a um resultado pouco confiável. O motivo para isso está no fato de poder haver um aumento em um determinado tipo de erro.

Realizando uma breve recapitulação, no experimento proposto tem-se que as unidades experimentais são os datasets; os tratamentos são os algoritmos de machine learning; e a variável resposta a ser trabalhada é o score de performance para a predição de cada modelo, sendo utilizado neste caso o R2.

## Testando normalidade

O trecho abaixo foi construído de modo a analisar se os dados trabalhados estavam atestando a normalidade. A partir de alguns testes e análise gráfica tem-se a verificação que os dados atendem a referida característica. Contudo, é importante comentar que a quantidade de amostras disponíveis para o experimento é baixa.

```python
pg.normality(data=df_res, dv='R2', group='model')
```

```python
import statsmodels.api as sm
from statsmodels.formula.api import ols

formula = 'R2 ~ model'

model = ols(formula, data=df_res).fit()
res = model.resid
```

```python
# QQ plot
ax = pg.qqplot(res, dist='norm')
```

```python
import seaborn as sns

ax = sns.histplot(res)
```

```python
#Shapiro-Wilk test with scipy

from scipy.stats import shapiro

shapiro(res)
```

## Testando Sphericity

No que se refere ao trecho seguinte, ele trata de verificar o atributo de Sphericity para os dados, sendo utilizado o teste de Mauchly. Quanto ao resultado, tem-se que os dados atendem a dada característica.

```python
pg.sphericity(data=df_res, dv='R2', subject='dataset', within='model')
```

## ANOVA

Por fim, o resultado da ANOVA com medidas repetidas aponta que não devemos rejeitar a hipótese nula, ou seja, não há diferença entre os algoritmos de machile learning para a predição dos datasets. Contudo, importante apontar que o p valor encontrado localiza-se bem perto do limiar do nível de significância.Além disso, por experiências práticas acredita-se que  existam algoritmos melhores para determinados contextos, considerando isso, talvez com um número maior de amostras o resultado encontrado difera daquele obtido para ese trabalho.
